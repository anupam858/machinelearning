```python
import numpy as np
import pandas as pd
```

```python
X = pd.read_csv('D:/college/MachineLearningTraining/train.csv')
```

```python
X.head()
```

```python
X.tail()
```

```python
X["Age"].fillna(X.groupby("Sex")["Age"].transform("median"), inplace=True)
```

```python
def bar_chart(feature):
    survived = X[X['Survived']==1][feature].value_counts()
    dead = X[X['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['Survived','Dead']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
bar_chart('Sex')
bar_chart('Pclass')
bar_chart('SibSp')
bar_chart('Embarked')
bar_chart('Parch')
```

```python
Y = X['Survived']
X = X.drop('Survived', axis=1)

X.head()
```

```python
dummy_variables = pd.get_dummies(X[['Pclass','Sex','Embarked']])
```

```python
dummy_variables
```

```python
dummy_variables['Fare'] = X['Fare'].values
dummy_variables['Age'] = X['Age'].values
```

```python
dummy_variables.head(5)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(dummy_variables, Y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression().fit(X_train,Y_train)
```

```python
Y_pred = lg.predict(X_test)
```

```python
print(Y_test.values)

print(Y_pred)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
dtc = DecisionTreeClassifier(max_depth=5).fit(X_train,Y_train)

Y_pred_dtc = dtc.predict(X_test)

print(confusion_matrix(Y_test, Y_pred_dtc))
print(classification_report(Y_test, Y_pred_dtc))
```

```python
from graphviz import Source
```

```python
dot_data = export_graphviz(dtc, out_file=None)
graph = Source(dot_data)

graph.render("decision graph 2")
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB().fit(X_train, Y_train)

Y_pred_gnb = gnb.predict(X_test)
```

```python
print(classification_report(Y_test, Y_pred_gnb))
```

# Support Vector Machine

```python
from sklearn.svm import SVC

svc = SVC(C= 4,kernel = 'linear').fit(X_train, Y_train)

Y_pred_svc = svc.predict(X_test)

print(classification_report(Y_test, Y_pred_svc))
```
