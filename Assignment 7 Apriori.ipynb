# Apriori

```python
dataset = pd.read_csv('D:\college\MachineLearningTraining\Retail_Data.csv')
```

```python
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
X = pd.get_dummies(dataset[['Product1','Product2','Product3']])
```

```python
X.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset[['Product1', 'Product2', 'Product3']].values.tolist()).transform(dataset[['Product1', 'Product2', 'Product3']].values.tolist())
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df.head()
```

```python
from mlxtend.frequent_patterns import apriori

apriori(X, min_support=0.2, use_colnames =True)
```

# K- means Clustering

```python
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
```

```python
k_means = KMeans(n_clusters=10).fit(X)
```

```python
k_means.cluster_centers_
```

```python
k_means.labels_
```

```python
X['clusters'] = k_means.labels_
```
